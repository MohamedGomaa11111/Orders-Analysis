```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv (
   r"c:\Users\mohamed\Downloads\Orders (1).csv" ,
    skiprows=4
)
```

```python
df.dropna(how='all',inplace=True)
```

```python
df['City']=df['City'].str.title()
```

```python
df.head()
```

```python
df['Branch']=df['Branch'].str.upper()
df['Email']=df['Email'].str.lower()
```

```python
df['Phone Number']=df['Phone Number'].str.replace('Tel: ','')
```

```python
df['Discount']=df['Discount'].round(0)
```

```python
df['Total Price']=df['Quantity']*df['Unit Price']
```

```python
df['cost']=df['Total Price']-df['Discount'].fillna(0)
```

```python
df
```

```python
df.insert(9,'Customar Full Name',df['Customer First Name']+' '+df['Customer Last Name'])
```

```python
df
```

```python
newColom=df['Branch'].str[:2]
df.insert(5,'Br_code',newColom)

```

```python
df.loc[df['Status']==True,'order_status']='Done'
df.loc[df['Status']==False,'order_status']='pending'
```

```python
df
```

```python
def level(row):
  if row >1000:
    return 'high'
  else:
   return 'low'
df['cost_level']=df['cost'].apply(level)

```

```python
df['discount_discription']=np.where(df['Discount'].isnull(),'nodiscount',df['Discount']/df['cost'])
```

```python
sales_by_category = df.groupby("Category")["Total Price"].sum().sort_values(ascending=False)
```

```python


plt.figure(figsize=(10,6))

sales_by_category.plot(
    kind="bar",
    color="steelblue",
    edgecolor="black"
)

plt.title("Sales by Category", fontsize=14, fontweight='bold')
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
Best_Selling_Items=df['Item'].value_counts().head(5)
Best_Selling_Items
```

```python





plt.figure(figsize=(10,6))

colors = plt.cm.Blues(Best_Selling_Items / Best_Selling_Items.max())

Best_Selling_Items.plot(
    kind="bar",
    color=colors,
    edgecolor="black"
)

plt.title("Top 5 Selling Items", fontsize=14, fontweight='bold')
plt.xlabel("Item", fontsize=12)
plt.ylabel("Number of Sales", fontsize=12)

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
Status_count=df['order_status'].value_counts()
Status_count
```

```python



plt.figure(figsize=(7,7))

explode = [0.05] + [0]*(len(Status_count)-1)

Status_count.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=["#4C72B0", "#55A868", "#C44E52", "#8172B2"],
    wedgeprops={"edgecolor": "white", "linewidth": 2},
    explode=explode,
    shadow=True
)

plt.title("Order Status Distribution", fontsize=14, fontweight='bold')
plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
pivot = pd.pivot_table(
    df,
    values="Total Price",
    index="Category",
    columns="order_status",
    aggfunc="sum",
    fill_value=0
)

pivot
```

```python
pivot.plot(kind="bar", figsize=(10,6))


plt.title("Sales by Category and Status")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
pivot2=pd.pivot_table(
    df,
    values="Total Price",
    index="City",
    columns="Category",
    aggfunc="sum",
    fill_value=0
)
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    pivot2,
    annot=True,
    fmt=".0f",
    cmap="Blues"
)

plt.title("Sales Heatmap: City vs Category")
plt.show()
```

```python
sales_by_city_status = pd.pivot_table(
    df,
    values="Total Price",
    index="City",
    columns="order_status",
    aggfunc="sum",
    fill_value=0
)
sales_by_city_status 
```

```python


ax = sales_by_city_status.plot(
    kind="bar",
    stacked=True,
    figsize=(11,6),
    colormap="Set2",
    edgecolor="white",
    linewidth=0.7
)

plt.title("Sales by City and Status", fontsize=14, fontweight='bold')
plt.xlabel("City")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.legend(title="Status", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

```python
category_sales = pd.pivot_table(
    df,
    values="Total Price",
    index="Category",
    aggfunc="sum"
)
total_sales = category_sales["Total Price"].sum()
category_sales["Contribution %"] = (
    category_sales["Total Price"] / total_sales
) * 100
category_sales
```

```python

plt.figure(figsize=(7,7))

colors = ["#66b3ff", "#99ff99", "#ff9999", "#ffcc99", "#c2c2f0"]

category_sales["Contribution %"].head(5).plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    pctdistance=0.75,
    labeldistance=1.1,
    wedgeprops={"edgecolor": "white", "linewidth": 2}
)

plt.title("Category Contribution %", fontsize=14, fontweight='bold')
plt.ylabel("")

plt.tight_layout()
plt.show()
```
